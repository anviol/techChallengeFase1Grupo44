#   <font color=cyan> DF Quantidade - Exportação

```python
df = pd.DataFrame(dados_vinho)
```

```python
# Renomeando tabela de QUANTIDADE - Sabe-se que 1 Kg = 1L
df.rename(columns={'2008': '2008(L)', '2009': '2009(L)','2010': '2010(L)','2011': '2011(L)','2012': '2012(L)','2013': '2013(L)','2014': '2014(L)','2015': '2015(L)','2016': '2016(L)','2017': '2017(L)','2018': '2018(L)','2019': '2019(L)','2020': '2020(L)','2021': '2021(L)','2022': '2022(L)'}, inplace=True)
```

```python
# DF Quantidade em Litros - Sabe-se que 1 Kg = 1L
dados_exp_qtd = ['País','2008(L)', '2009(L)', '2010(L)', '2011(L)', '2012(L)', '2013(L)', '2014(L)', '2015(L)', '2016(L)', '2017(L)', '2018(L)', '2019(L)', '2020(L)', '2021(L)', '2022(L)']
df_qtd = df[dados_exp_qtd]
df_qtd
```

#   <font color=cyan> DF Valores - Exportação

```python
# Renomeando tabela de EXPORTAÇÃO do Valor
df.rename(columns={'2008.1': '2008(US$)', '2009.1': '2009(US$)','2010.1': '2010(US$)','2011.1': '2011(US$)','2012.1': '2012(US$)','2013.1': '2013(US$)','2014.1': '2014(US$)','2015.1': '2015(US$)','2016.1': '2016(US$)','2017.1': '2017(US$)','2018.1': '2018(US$)','2019.1': '2019(US$)','2020.1': '2020(US$)','2021.1': '2021(US$)','2022.1': '2022(US$)'}, inplace=True)
```

```python
# DF Valor de Exportação (Dollar - US$)
dados_exp_valor = ['País','2008(US$)', '2009(US$)','2010(US$)','2011(US$)','2012(US$)','2013(US$)','2014(US$)','2015(US$)','2016(US$)','2017(US$)','2018(US$)','2019(US$)','2020(US$)','2021(US$)','2022(US$)']
df_valor = df[dados_exp_valor]
df_valor.head()
```

#   <font color=cyan>  Renomeando INDEX

```python
df_qtd = df_qtd.set_index('País')
df_qtd
```

```python
df_valor = df_valor.set_index('País')
df_valor
```

#   <font color=cyan>  Removendo Duplicados e Sem valor

```python
# Removendo linhas desnecessárias e sem Valor - Qtd. Litros
df_qtd.drop_duplicates(inplace = True)
display(df_qtd)
```

```python
# Removendo linhas desnecessárias e Sem Valor - Valor em Dollar
df_valor.drop_duplicates(inplace = True)
display(df_valor)
```

#   <font color=cyan>  Soma Total - Linhas

```python
# Soma Qtd. em Litros
df_qtd["Soma"] = df_qtd.sum(axis=1)
df_qtd
```

```python
# Soma Valor em Dollar
df_valor["Soma"] = df_valor.sum(axis=1)
df_valor
```

#   <font color=cyan>  Fazendo Transposição para importar PBI

```python
# Transponha o DataFrame
df_qtd_transposto = df_qtd.transpose()

df_qtd_transposto.index.names = ['Ano'] # Nome Index

df_qtd_transposto.columns.names = ['']  # Nome Colunas

df_qtd_transposto
```

```python
df_valor_final = pd.DataFrame(df_valor)

# Transponha o DataFrame
df_valor_transposto = df_valor_final.transpose()

df_valor_transposto.index.names = ['Ano'] # Nome Index

df_valor_transposto.columns.names = ['']  # Nome Colunas

df_valor_transposto
```

# <font color=cyan> Lista de População

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL da página da Wikipedia com a tabela
url = 'https://pt.wikipedia.org/wiki/Lista_de_países_por_população'

# Fazendo a requisição para a página
response = requests.get(url)

# Criando o objeto BeautifulSoup para analisar o HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrando a tabela desejada (pode precisar inspecionar o HTML da página para encontrar o nome correto da classe ou do ID)
table = soup.find('table', {'class': 'wikitable sortable'})

# Lendo a tabela usando o pandas
df = pd.read_html(str(table))[0]

# Exibindo as primeiras linhas da tabela
print(df.head())
```

```python
populacao_paises = df[['País (ou território dependente)','Estimativa da ONU']]

paises_interesse = ['Brasil', 'Rússia', 'Estados Unidos', 'França', 'Itália', 'Alemanha', 'China', 'Reino Unido', 'Espanha', 'Argentina', 'Austrália', 'Portugal', 'Canadá', 'Romênia', 'África do Sul', 'Países Baixos', 'Bélgica', 'Japão']
populacao_paises_temp = populacao_paises[populacao_paises['País (ou território dependente)'].isin(paises_interesse)]
```
